```python
# Import Dependencies
import pandas as pd
```

```python
# Create reference to CSV file
file_path = "../Resources/SFO_Airport_Utility_Consumption.csv"
# Import the CSV into a pandas DataFrame
sfo_utlity_consumptions_df = pd.read_csv(file_path)
sfo_utlity_consumptions_df
```

```python
# Collect a list of all the unique values in "Utility"
unique_utilities_df = sfo_utlity_consumptions_df["Utility"].unique()
unique_utilities_df
```

```python
# Create a new DataFrame that only includes electricity usage for tenant owned units.

tenant_electricity_df = sfo_utlity_consumptions_df.loc[sfo_utlity_consumptions_df["Utility"] == "Electricity"]
tenant_electricity_df
```

```python
# Sort the DataFrame by the values in the "Usage" column to find the months
# with the highest usage
usage_sorted_df = tenant_electricity_df.sort_values("Usage", ascending=False)
# Reset the index so that the index is now based on the sorting locations
usage_sorted_df = usage_sorted_df.reset_index(drop=False)
usage_sorted_df
```

```python
# Save all of the information collected on the highest usage month
highest_usage_month_df = usage_sorted_df.iloc[0]["Month"]
highest_usage_month_df
```

```python
# Show some basic summary statistics for the whole DataFrame
summary_statistics_df = sfo_utlity_consumptions_df.describe()
summary_statistics_df
```

```python
# Find the average usage
average_usage_df = sfo_utlity_consumptions_df["Usage"].mean()
average_usage_df
```

```python
# Find the total usage from all Augusts on record
total_usage_august_df = sfo_utlity_consumptions_df.loc[sfo_utlity_consumptions_df["Month"] == "Aug"]["Usage"].sum()
total_usage_august_df
```
